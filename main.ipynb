```python
import os
import json
import csv
import time
import logging
import urllib.request
import urllib.error
# Logging setup
logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s")
# API configuration
BASE_URL = "https://api.pagepixels.com/snap"
ACCESS_TOKEN = "satyrkjDNv2C6HABRsQtyoY6ojChKL1_6l8cUXw3NLQ"
# Retry settings
MAX_RETRIES = 3  
OUTPUT_FILE = "url_snapshots.csv"  

def get_request_url(url):
    """Build the API request URL with parameters."""
    params = {
        "ttl": 1500,
        "url": url,
        "access_token": ACCESS_TOKEN,
        "json": "true",
        "no_cookie_banners": "true",
        "no_ads": "true",
        "time_zone": "US/Eastern",
    }
    return f"{BASE_URL}?{'&'.join(f'{k}={v}' for k, v in params.items())}"

def fetch_snapshot(url):
    """Fetch a snapshot for a given URL with retry mechanism (sequentially)."""
    request_url = get_request_url(url)
    retries = 0

    while retries < MAX_RETRIES:
        try:
            with urllib.request.urlopen(request_url) as response:
                if response.status != 200:
                    logging.error(f"Error {response.status} for {url}")
                    return url, None
                
                response_data = response.read().decode("utf-8")
                json_response = json.loads(response_data)
                direct_link = json_response.get("direct_link")

                if not direct_link:
                    logging.warning(f"Unexpected response for {url}: {json_response}")
                    return url, None

                logging.info(f"Snapshot succeeded for {url}: {direct_link}")
                return url, direct_link

        except (urllib.error.HTTPError, urllib.error.URLError) as e:
            logging.error(f"Error fetching {url}: {e}")
        except json.JSONDecodeError:
            logging.error(f"Invalid JSON response for {url}")
        
        retries += 1
        time.sleep(2**retries)  # Exponential backoff

    return url, None

def save_result(output_file, url, direct_link, is_new_file):
    """Save each result immediately to the CSV file (line by line)."""
    try:
        with open(output_file, "a", newline="") as output_csv:
            fieldnames = ["url", "direct_link"]
            writer = csv.DictWriter(output_csv, fieldnames=fieldnames)

            if is_new_file:
                writer.writeheader()

            writer.writerow({"url": url, "direct_link": direct_link})

    except Exception as e:
        logging.error(f"Error writing to CSV: {e}")

def process_urls(url_list):
    """Process the list of URLs **sequentially** and save each result immediately."""
    is_new_file = not os.path.exists(OUTPUT_FILE)

    for url in url_list:
        url, direct_link = fetch_snapshot(url)
        if direct_link:
            save_result(OUTPUT_FILE, url, direct_link, is_new_file)
            is_new_file = False  # Ensure headers are written only once

    logging.info(f"Processing complete. Results saved to {OUTPUT_FILE}")
process_urls(url_list)
```
