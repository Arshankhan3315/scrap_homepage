```python
import csv
import os
import json
import urllib.request
import urllib.error
import concurrent.futures

# File paths
input_file_path = os.path.expanduser("/content/web1_url - Sheet1.csv")
out_file_path = input_file_path.replace(".csv", "_results.csv")

# API configuration
BASE_URL = "https://api.pagepixels.com/snap"
ACCESS_TOKEN = "satyrkjDNv2C6HABRsQtyoY6ojChKL1_6l8cUXw3NLQ"

# Concurrency settings
MAX_CONCURRENT_REQUESTS = 10
test_count = 1  # Limit the number of rows processed during testing. 0 processes all rows.
start_after_url = ""  # Skip rows before this URL

def get_request_url(url):
    params = {
        "ttl": 1500,
        "url": url,
        "access_token": ACCESS_TOKEN,
        "json": "true",
        "no_cookie_banners": "true",
        "no_ads": "true",
        "time_zone": "US/Eastern",
    }
    return f"{BASE_URL}?{'&'.join(f'{k}={v}' for k, v in params.items())}"

def fetch_snapshot(url):
    """Fetch snapshot for a given URL and return the result."""
    request_url = get_request_url(url)
    try:
        with urllib.request.urlopen(request_url) as response:
            if response.status != 200:
                print(f"Error for {url}: HTTP {response.status}")
                return url, None

            response_data = response.read().decode('utf-8')
            json_response = json.loads(response_data)
            direct_link = json_response.get('direct_link')

            if not direct_link:
                print(f"Unexpected response for {url}: {json_response}")
                return url, None

            print(f"Snapshot for {url} succeeded: {direct_link}")
            return url, direct_link
    except urllib.error.HTTPError as e:
        print(f"HTTP error for {url}: {e.code} {e.reason}")
    except urllib.error.URLError as e:
        print(f"URL error for {url}: {e.reason}")
    except json.JSONDecodeError:
        print(f"Invalid JSON response for {url}")

    return url, None

def save_result(output_file_path, url, direct_link, is_new_file):
    """Appends a single result to the CSV file after each API call."""
    try:
        with open(output_file_path, 'a', newline='') as output_csv:
            fieldnames = ['url', 'direct_link']
            writer = csv.DictWriter(output_csv, fieldnames=fieldnames)

            if is_new_file:
                writer.writeheader()

            writer.writerow({'url': url, 'direct_link': direct_link})
    except Exception as e:
        print(f"Error writing result for {url}: {e}")

def process_csv(input_file_path, output_file_path, start_after_url, test_count):
    """Processes the input CSV with concurrent requests and saves results after each call."""
    is_new_file = not os.path.exists(output_file_path)
    urls_to_process = []

    try:
        with open(input_file_path, 'r') as csv_file:
            reader = csv.DictReader(csv_file)
            row_count = 0
            start_processing = start_after_url == ""

            for row in reader:
                url = row.get('url')
                print(row)
                if not url:
                    continue

                if not start_processing:
                    if url == start_after_url:
                        start_processing = True
                    continue

                urls_to_process.append(url)
                row_count += 1

                if test_count and row_count >= test_count:
                    break
    except FileNotFoundError:
        print(f"File not found: {input_file_path}")
        return
    except Exception as e:
        print(f"An error occurred: {e}")
        return

    with concurrent.futures.ThreadPoolExecutor(max_workers=MAX_CONCURRENT_REQUESTS) as executor:
        future_to_url = {executor.submit(fetch_snapshot, url): url for url in urls_to_process}
        for future in concurrent.futures.as_completed(future_to_url):
            url, direct_link = future.result()
            if direct_link:
                save_result(output_file_path, url, direct_link, is_new_file)
                is_new_file = False  # Ensure headers are written only once

    print(f"Processing complete. Results saved to {output_file_path}")

# Run the process
process_csv(input_file_path, out_file_path, start_after_url, test_count)
```
